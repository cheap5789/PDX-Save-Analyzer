Search for any key anywhere in the EU5 save JSON tree. Substring search by default, with optional exact match. Prints the full dot-path and value for each match.

```python
import sys
from pathlib import Path

# Get the absolute path of the current directory
current_dir = Path.cwd()

# If 'notebooks' is in the path, move to its parent
if current_dir.name == "notebooks":
    PROJECT_ROOT = current_dir.parent
else:
    # Assuming if we aren't in notebooks, we are already at root 
    # (or you can add logic to find the root)
    PROJECT_ROOT = current_dir

# Set the working directory
import os
os.chdir(PROJECT_ROOT)

# Add to sys.path
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

print(f"Project root: {PROJECT_ROOT}")
```

```python
# === CONFIGURATION ===
SAVE_FILE   = "saves/autosave.eu5"
RAKALY_BIN  = "bin/rakaly/rakaly"
```

```python
from toolbox.save_loader import load_save
from toolbox.find_key import _search, _format_value

save = load_save(SAVE_FILE, rakaly_bin=RAKALY_BIN, verbose=True)
data = save.raw
print("Save loaded.")
```

## Search

Edit `SEARCH_TERM`, `EXACT`, and `MAX_RESULTS` and re-run the cell.

```python
# === SEARCH PARAMETERS ===
SEARCH_TERM = "country_type"       # What to search for
EXACT       = False             # True = exact key match, False = substring
MAX_RESULTS = 10                # Stop after this many hits

count = 0
for path, val in _search(data, SEARCH_TERM, EXACT, []):
    print(f"  {path}")
    print(f"    = {_format_value(val)}")
    count += 1
    if count >= MAX_RESULTS:
        print(f"\n... stopped at {MAX_RESULTS} results.")
        break

if count == 0:
    print(f"No keys matching {SEARCH_TERM!r} found.")
else:
    print(f"\n{count} result(s) found.")
```

<h2>Search term with Pandas Table</h2>

```python
import pandas as pd
from IPython.display import display, HTML

# === SEARCH PARAMETERS ===
SEARCH_TERM = "CabinetCardModifier"
EXACT       = False
MAX_RESULTS = 1000 

results = []
for path, val in _search(data, SEARCH_TERM, EXACT, []):
    results.append({"Path": path, "Value": _format_value(val)})
    if len(results) >= MAX_RESULTS:
        break

if not results:
    print(f"No keys matching {SEARCH_TERM!r} found.")
else:
    df = pd.DataFrame(results)

    # 1. Prevent Pandas from truncating the list (shows all 2500 rows)
    pd.set_option('display.max_rows', None)
    
    # 2. Render the DataFrame to HTML
    # We use classes='table' for standard styling
    df_html = df.to_html(index=False)

    # 3. Wrap in a scrollable DIV (height 400px - change as needed)
    scroll_html = f"""
    <div style="height:400px; width:100%; overflow-y:scroll; border:1px solid #ccc;">
        {df_html}
    </div>
    <p><b>Total Results: {len(results)}</b></p>
    """
    
    display(HTML(scroll_html))
    
    # Optional: Reset option so other cells don't explode
    pd.reset_option('display.max_rows')
```

## Quick Searches

Run multiple searches in one go. Add terms to the list below.

```python
TERMS = ["gold", "manpower", "stability", "war_exhaustion", "prestige"]

for term in TERMS:
    results = list(_search(data, term, exact=True, path=[]))
    print(f"\n{'='*60}")
    print(f"  {term!r} — {len(results)} result(s)")
    print(f"{'='*60}")
    for path, val in results[:5]:
        print(f"  {path}")
        print(f"    = {_format_value(val)}")
```

## Collect Results Programmatically

```python
# Collect all matches for a key into a list for further processing
results = [(path, val) for path, val in _search(data, "gold", exact=True, path=[])]
print(f"Found {len(results)} matches for 'gold' (exact)")

# Example: extract all gold values from country currency_data
gold_values = [
    (path, val) for path, val in results
    if "currency_data" in path and isinstance(val, (int, float))
]
print(f"Of which {len(gold_values)} are in currency_data")
for path, val in gold_values[:10]:
    print(f"  {path}: {val}")
```
